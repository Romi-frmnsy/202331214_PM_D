```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.metrics import mean_squared_error, accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Memuat dataset
data_korupsi = pd.read_csv('R04_corruptions.csv')
data_diabetes = pd.read_csv('K02_diabetes.csv')
```

```python
# Menampilkan informasi dasar dari kedua dataset
print("Tampilan Awal Dataset Korupsi:")
print(data_korupsi.head())
print("\nTampilan Awal Dataset Diabetes:")
print(data_diabetes.head())
```

```python
# Untuk Dataset Korupsi (Regresi)
X_reg = data_korupsi.drop('Government Transparency (Y)', axis=1)
y_reg = data_korupsi['Government Transparency (Y)']
```

```python
# Untuk Dataset Diabetes (Klasifikasi)
X_clf = data_diabetes.drop('diabetes', axis=1)
y_clf = data_diabetes['diabetes']
```

```python
# Membagi dataset menjadi data latih dan data uji
X_latih_reg, X_uji_reg, y_latih_reg, y_uji_reg = train_test_split(X_reg, y_reg, test_size=0.3, random_state=42)
```

```python
# Inisialisasi dan melatih model Decision Tree Regressor
regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_latih_reg, y_latih_reg)
```

```python
# Membuat prediksi
y_pred_reg = regressor.predict(X_uji_reg)
```

```python
# Mengevaluasi model regresi
mse = mean_squared_error(y_uji_reg, y_pred_reg)
print(f"Mean Squared Error (Regresi): {mse}")
```

```python
# Menampilkan visualisasi Decision Tree untuk regresi
plt.figure(figsize=(20, 10))
tree.plot_tree(regressor, feature_names=X_reg.columns, filled=True)
plt.title("Decision Tree Regressor")
plt.show()
```

```python
X_latih_clf, X_uji_clf, y_latih_clf, y_uji_clf = train_test_split(X_clf, y_clf, test_size=0.3, random_state=42)
```

```python
# Mengonversi kolom kategorikal menjadi numerik menggunakan One-Hot Encoding
X_clf = pd.get_dummies(X_clf, columns=['gender', 'smoking_history'], drop_first=True)
```

```python
# Untuk Dataset Diabetes (Klasifikasi)
X_clf = data_diabetes.drop('diabetes', axis=1)
y_clf = data_diabetes['diabetes']
```

```python

# Mengonversi kolom kategorikal menjadi numerik
X_clf = pd.get_dummies(X_clf, columns=['gender', 'smoking_history'], drop_first=True)
```

```python
X_latih_clf, X_uji_clf, y_latih_clf, y_uji_clf = train_test_split(X_clf, y_clf, test_size=0.3, random_state=42)
```

```python

# Inisialisasi dan melatih model Decision Tree Classifier
classifier = DecisionTreeClassifier(random_state=42)
classifier.fit(X_latih_clf, y_latih_clf)
```

```python
# Membuat prediksi
y_pred_clf = classifier.predict(X_uji_clf)
```

```python
# Mengevaluasi model klasifikasi
akurasi = accuracy_score(y_uji_clf, y_pred_clf)
print(f"Akurasi (Klasifikasi): {akurasi}")
print("\nLaporan Klasifikasi:\n", classification_report(y_uji_clf, y_pred_clf))
```
